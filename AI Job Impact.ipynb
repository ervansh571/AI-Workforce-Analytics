```python
# Industry Vs Salary Before AI
# Industry Which Gives Highest Salary Before AI
# Industry Which Gives lowest Salary before AI
import pandas as pd
df = pd.read_csv('ai_job_impact.csv')
a= df.groupby("Industry")["Salary_Before_AI"].sum()
print(a)
b= df.groupby("Industry")["Salary_Before_AI"].sum().idxmax()
c=df.groupby("Industry")["Salary_Before_AI"].sum().max()
d= df.groupby("Industry")["Salary_Before_AI"].sum().idxmin()
e=df.groupby("Industry")["Salary_Before_AI"].sum().min()
print("Industry Which gives highest salary before Ai is:",b,"\nTotal Salary Given by that Industry to all Employees:",c)
print("Industry Which gives lowest salary before Ai is:",d,"\nTotal Salary Given by that Industry to all Employees:",e)
```

```python
# Industry Vs Salary After AI
# Industry Which Gives Highest Salary After AI
# Industry Which Gives lowest Salary After AI
import pandas as pd
df = pd.read_csv('ai_job_impact.csv')
a= df.groupby("Industry")["Salary_After_AI"].sum()
print(a)
b= df.groupby("Industry")["Salary_After_AI"].sum().idxmax()
c=df.groupby("Industry")["Salary_After_AI"].sum().max()
d= df.groupby("Industry")["Salary_After_AI"].sum().idxmin()
e=df.groupby("Industry")["Salary_After_AI"].sum().min()
print("Industry Which gives highest salary After Ai is:",b,"\nTotal Salary Given by that Industry to all Employees:",c)
print("Industry Which gives lowest salary After Ai is:",d,"\nTotal Salary Given by that Industry to all Employees:",e)
```
